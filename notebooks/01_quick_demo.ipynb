# Quick demo: Mixed-Precision IR
Solve a Poisson system and inspect telemetry.

```python
import numpy as np, scipy.sparse as sp
from mpir_sparse.ir import solve
nx, ny = 80, 80
N = nx*ny
main = 4.0*np.ones(N)
ew = -1.0*np.ones(N-1); ew[np.arange(1,N)%nx==0] = 0.0
ns = -1.0*np.ones(N-nx)
A = sp.diags([main, ew, ew, ns, ns],[0,-1,1,-nx,nx], format='csr', dtype=np.float64)
x_true = np.ones(N)
b = A @ x_true
x, info = solve(A, b, tol=1e-12, k_max=10, inner_tol=1e-3, work_dtype=np.float32, preconditioner='ilu')
info['converged'], info['final_backward_error'], info['escalations']
```
